## Funciones

````python
datos["Departamento"].unique()
````

````python
datos["Cod DANE"] = datos["Cod DANE"].astype(str)

datos.describe()
````

````python
datos["2001"] = datos["2001"].astype(float)
````

````python
datos["2001"] = datos["2001"].apply(lambda x : x.replace(",",".")).astype(float)
datos["2002"] = datos["2002"].apply(lambda x : x.replace(",",".")).astype(float)
datos["2003"] = datos["2003"].apply(lambda x : x.replace(",",".")).astype(float)
datos["2004"] = datos["2004"].apply(lambda x : x.replace(",",".")).astype(float)
datos["2005"] = datos["2005"].apply(lambda x : x.replace(",",".")).astype(float)
datos["2006"] = datos["2006"].apply(lambda x : x.replace(",",".")).astype(float)
datos.describe()
````

````python
columnas = [f"200{i}" for i in range(1,7)]

def procesar(x):
    if x.name in columnas:
        new_columna = x.apply(lambda x : x.replace(",","."))
        return new_columna.astype(float)
    else:
        return x

datosMod = datos.apply(procesar)
datosMod.describe()
````
